# WordCloud
Un nuage de mots (ou word cloud) est une représentation visuelle de la fréquence des mots dans un texte. Il peut être créé en Python avec la bibliothèque WordCloud, souvent utilisée pour analyser et visualiser la fréquence des mots dans des textes ou des colonnes de texte d'un DataFrame.


Utiliser la fonction ci-dessous pour faire le cloud des termes pertinents.

```python
from wordcloud import WordCloud


def wordcloud_fig (text):

  text = " ".join(text)
  wordcloud = WordCloud(width=800, height = 800,background_color= 'white').generate(text)
  return wordcloud
```
